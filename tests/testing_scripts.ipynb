# Here I am going to test different things

Here I test whether the Wesfeiler-Lehman Hashing works as I expect it to work. I create different graphs that should (or shold not) be equivalent and check if their hashing is the same. 

```python
# from population_graph import PopulationGraph
import networkx as nx

# Here I will test whether the WL is identical for two different graphs that are technically equivalent. 
# I will generate two circular graphs and will remove one edge (in different positions) and see if it gets identical

# Create two circular graphs (cycles)
n = 6  # number of nodes
G1 = nx.cycle_graph(n)
G2 = nx.cycle_graph(n)

# Remove one edge from each graph at different positions
# For G1, remove edge between nodes 0 and 1
G1.add_edge(4, 1)
# G1.remove_edge(4, 5)

# For G2, remove edge between nodes 2 and 3
G2.add_edge(2, 5)
G2.remove_edge(5, 4)

# Both graphs should now be paths and should be isomorphic
print("Graph G1 edges:", list(G1.edges()))
print("Graph G2 edges:", list(G2.edges()))

# Test if they are isomorphic
print("Are G1 and G2 isomorphic?", nx.is_isomorphic(G1, G2))

# Apply Weisfeiler-Lehman test
from networkx.algorithms import isomorphism

# Get WL hash for both graphs
wl_hash_G1 = isomorphism.categorical_node_match('label', '')
wl_hash_G2 = isomorphism.categorical_node_match('label', '')

# Alternative: use graph_hash for WL-like comparison
hash_G1 = nx.weisfeiler_lehman_graph_hash(G1)
hash_G2 = nx.weisfeiler_lehman_graph_hash(G2)

print(f"WL hash for G1: {hash_G1}")
print(f"WL hash for G2: {hash_G2}")
print(f"WL hashes are identical: {hash_G1 == hash_G2}")


# 2. Graph6 (Reversible Storage) -> DIFFERENT
print(f"Graph6 G1:  {nx.to_graph6_bytes(G1, header=False)}")
print(f"Graph6 G2:  {nx.to_graph6_bytes(G2, header=False)}")
```
